```python
import biotite.structure as struc
import biotite.structure.info as info
import ammolite
from pymol2 import PyMOL


PNG_SIZE = (600, 600)


pymol_app = PyMOL()
pymol_app.start()
ammolite.setup_parameters(pymol_instance=pymol_app)
cmd = pymol_app.cmd
```

```python
caffeine = info.residue("CFF")
#caffeine.chain_id[:] = "A"
#caffeine.res_id[:] = 1

pymol_obj = ammolite.PyMOLObject.from_structure(caffeine, pymol_instance=pymol_app)
ammolite.show(PNG_SIZE, pymol_instance=pymol_app)
```

```python
cmd.orient()
cmd.zoom(buffer=2.0)
ammolite.show(PNG_SIZE, pymol_instance=pymol_app)
```

```python
pymol_obj.set("stick_radius", 0.15)
pymol_obj.show("spheres")
pymol_obj.set("sphere_scale", 0.25)
pymol_obj.set("sphere_quality", 4)
pymol_obj.set("stick_color", "grey80")
pymol_obj.color("black", caffeine.element == "C")
ammolite.show(PNG_SIZE, pymol_instance=pymol_app)
```

```python
pymol_app.stop()
```
